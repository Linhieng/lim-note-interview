## 1. 简单使用

和条形图类似, 但柱状图一般是竖着放的。

````python
# 数据
labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 23, 45, 32, 18]

# 绘制垂直条形图
plt.bar(labels, values)

# 添加标题和标签
plt.title('Bar Chart Horizontal')
plt.xlabel('Category')
plt.ylabel('Value')

# 显示图表
plt.show()
````

````python
# 数据
data = [(0, 0.174), (1, 0.088), (2, 0.086), (3, 0.08),
        (4, 0.068), (5, 0.058), (6, 0.062), (7, 0.384)]

# 将数据分为横坐标和纵坐标
x = [d[0] for d in data]
y = [d[1] for d in data]

# 创建一个新的图像，并指定宽度为 6 英寸, 高度为 4 英寸
plt.figure(figsize=(6, 4))

# 绘制直方图
plt.bar(x, y, width=0.4)  # 后面的参数是可选的

# 在每个柱子上显示纵坐标的值，旋转角度为 90 度
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.02,
             '{:.3f}'.format(y[i]), rotation=90, ha='center', fontsize=14)

# 设置横纵坐标的刻度
plt.xticks(range(8))
plt.yticks([i/10 for i in range(6)])

# 显示图形
plt.show()
````

说明:
```py
plt.text(x[i] + 0.25, y[i]/2, '{:.3f}'.format(y[i]), rotation=90, ha='center', fontsize=8)
```
- `x[i] + 0.25` 表示文本放置的横坐标位置。这里加上 0.25 的偏移量是为了让文本居中显示在柱子的上方。
- `y[i]/2` 表示文本放置的纵坐标位置。这里使用 y[i]/2 是为了将文本放置在柱子高度的一半处。
- `'{:.3f}'.format(y[i])` 表示要显示的文本内容，即纵坐标的值。{:.3f} 表示格式化字符串，表示保留小数点后三位。
- `rotation=90` 表示旋转角度，这里设置为 90 度，使文本垂直显示。
- `ha='center'` 表示水平对齐方式，这里设置为居中对齐，使文本水平居中显示。
- `fontsize=8` 表示字体大小，这里设置为 8 号字体。

````python
import pandas as pd
import matplotlib.pyplot as plt

# 创建数据
data = {'Class': [0, 0, 1] }
df = pd.DataFrame(data)

# 统计 Class 值中 0 和 1 出现的次数
counts = df['Class'].value_counts()

# 绘制柱状图
ax = counts.plot(kind='bar', rot=45, xlabel='Class', ylabel='Count')

# 在柱形上添加数值标签
for p in ax.containers:
    ax.bar_label(p, label_type='edge')

# 展示图形
plt.show()
````
