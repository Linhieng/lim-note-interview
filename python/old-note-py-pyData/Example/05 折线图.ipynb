## 2. 介绍 plot()

`plt.plot()` 是 Matplotlib 中用于绘制折线图的函数。它的基本语法如下:

```py
plt.plot(x, y, fmt, **kwargs)
```

- `x` 和 `y` 分别是用于绘制折线图的数据, 可以是列表、Numpy 数组或 Pandas Series 对象; 
- `fmt` 用于定义线条的样式字符串。 这个字符串包含一个或多个单字符代码，每个代码控制一种属性，例如颜色、线型和标记类型等
- `kwargs` 是一个字典, 用于指定其他绘图参数, 例如标题、坐标轴标签和注释等。

`fmt` 控制颜色的字符串有:

- `'b'`: 蓝色
- `'g'`: 绿色
- `'r'`: 红色
- `'c'`: 青色（蓝绿色）
- `'m'`: 洋红色（品红色）
- `'y'`: 黄色
- `'k'`: 黑色
- `'w'`: 白色

`fmt` 控制线条形状的字符串有:

- `'-'`: 实线（默认）
- `'--'`: 虚线
- `'-.'`: 点划线
- `':'`: 点线
- `'_'`: 水平线条

`fmt` 控制每个点标记类型的字符串有:

- `'.'`: 小圆点
- `','`: 像素点标记
- `'o'`: 圆圈标记
- `'v'`: 倒三角形标记
- `'^'`: 正三角形标记
- `'<'`: 左三角形标记
- `'>'`: 右三角形标记
- `'1'`: 下花心标记
- `'2'`: 上花心标记
- `'3'`: 左花心标记
- `'4'`: 右花心标记
- `'s'`: 正方形标记
- `'p'`: 五边形标记
- `'*'`: 星号标记
- `'h'`: 六边形1标记
- `'H'`: 六边形2标记
- `'+'`: 加号标记
- `'x'`: 叉号标记
- `'D'`: 菱形标记
- `'d'`: 小菱形标记

上面介绍的三类 `fmt` 可以组合使用, 比如 `'r--H'`。

其实 `fmt` 是一种简便的写法, 也可以不使用 `fmt`, 而是使用 `kwargs` 来定义。
常见的 `kwargs` 有:

- `color`: 线条颜色
- `linestyle`, `ls`: 线型
  - `linewidth`, `lw`: 线宽
- `marker`: 标记类型
  - `markersize`, `ms`: 标记大小
  - `markeredgewidth`: 标记宽度
  - `markeredgecolor`: 标记颜色
- `label`: 添加一个图例标签，用于区分不同的曲线。

````python
# 随机生成数据
x = list(range(10))
y = [random.randint(1, 10) for i in x]

# 绘制折线图
plt.plot(x, y, 'r--o')

# 设置图形标题和坐标轴标签
plt.title("Example Line Plot")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")

# 显示图形
plt.show()
````

### 绘制多条曲线

````python
import numpy as np
````

````python
# 生成 x 坐标轴数据
x = np.linspace(0, 2*np.pi, 100)

# 生成 y 坐标轴数据
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制两条曲线
plt.plot(x, y1, label='sin 曲线')
plt.plot(x, y2, label='cos 曲线')

# 添加标题和图例
plt.title('Sine and Cosine')
plt.legend()

# 显示图形
plt.show()
````
