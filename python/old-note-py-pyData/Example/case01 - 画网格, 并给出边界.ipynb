```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 创建一个 65 * 4 的矩阵, 每个单元格对应划分的网格, 单元格内有四个值, 分别对应网格的上下左右边界值

grid_margin = np.zeros((13*5, 4))

margin_top = np.repeat(range(17, 26, 2), 13)
margin_bottom = np.repeat(range(15, 24, 2), 13)
margin_left = np.tile(range(100, 125, 2), 5)
margin_right = np.tile(range(102, 127, 2), 5)

grid_margin[:, 0] = margin_top
grid_margin[:, 1] = margin_bottom
grid_margin[:, 2] = margin_left
grid_margin[:, 3] = margin_right

grid_margin
```

```python
# 创建一个 20 x 8 大小的画布
fig = plt.figure(figsize=(20, 8))

# 将画布划分成 5 行 13 列的网格子区域， 并获取每个子区域的Axes对象
axes = []
for row in range(4, -1, -1):
    for col in range(13):
        ax = fig.add_subplot(5, 13, row * 13 + col + 1, aspect='equal')
        ax.set_xticks([])  # 清空x轴的刻度值
        ax.set_yticks([])  # 清空y轴的刻度值
        ax.set_title((col+1) + 13*(4 - row)  , fontsize=20, y=0.3)
        axes.append(ax)

# 对每个子区域的Axes对象进行设置和绘图
for i, ax in enumerate(axes):
    ax.text(0.5, 0.95, int(grid_margin[i][0]), ha='center', va='top', fontsize=12, transform=ax.transAxes)
    ax.text(0.5, 0.05, int(grid_margin[i][1]), ha='center', va='bottom', fontsize=12, transform=ax.transAxes)
    ax.text(0.05, 0.5, int(grid_margin[i][2]), ha='left', va='center', fontsize=12, rotation=90, transform=ax.transAxes)
    ax.text(0.95, 0.5, int(grid_margin[i][3]), ha='right', va='center', fontsize=12, rotation=-90, transform=ax.transAxes)

plt.show()
```
