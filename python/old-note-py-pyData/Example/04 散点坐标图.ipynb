## 排列组合各列, 并依次绘制其散点图(更灵活)

````python
def scatter_matrix(dataframe):
    # 确定数据集中列的数量
    num_cols = len(dataframe.columns)
    
    # 创建散点图矩阵
    for i in range(num_cols):
        for j in range(i+1, num_cols):
            fig = plt.figure()
            ax = fig.add_subplot(111)
            ax.scatter(dataframe.iloc[:, i], dataframe.iloc[:, j])
            ax.set_xlabel(dataframe.columns[i])
            ax.set_ylabel(dataframe.columns[j])
            plt.show()
````

````python
# 调用scatter_matrix函数
scatter_matrix(data)
````

## 使用 pandas 自带的绘制散点图矩阵(示例)

scatter_matrix `是一个用于绘制多个变量之间散点图矩阵的函数，其主要参数如下`: 

- frame: DataFrame对象，包含待绘制散点图的数据
- alpha: 点透明度，取值范围为[0, 1]
- figsize: 图像尺寸，形式为元组，例如(10, 10)
- diagonal: 设置对角线上的单变量分布图，可以选择'hist'或'kde'
- marker: 散点的标记样式，默认为'o'，还可以使用其他的标记样式，例如"."、"+"等
- density_kwds: 字典类型，传递给核密度估计图(kde)的参数设置
- hist_kwds: 字典类型，传递给直方图(hist)的参数设置
- range_padding: 控制图像中每个小图与边界的距离
- label: 可以是字符串或数组，包含每个小图的标签名称
- color: 指定散点的颜色，可以是单一整数或字符串，也可以通过传递列表来指定每个数据集的颜色
需要注意的是，scatter_matrix 默认会针对每个特征之间的组合绘制散点图，因此如果传入的特征数量较多，那么将会产生大量图像，可能会导致绘图效率和可读性降低。因此在实际使用时需要根据数据集的大小和特征数量进行选择和调整。

语法:
```py
pandas.plotting.scatter_matrix(frame, alpha=0.5, figsize=None, ax=None, diagonal='hist', marker='.', density_kwds=None, hist_kwds=None, range_padding=0.05, **kwargs)
```
其中，必选参数 `frame` 表示待绘制散点图的数据集。

可选参数包括: 

- `alpha`: 点透明度，默认为 0.5。
- `figsize`: 绘图尺寸，形式为元组，例如 (10, 10)。
- `ax`: 指定绘图使用的坐标轴对象，如果没有指定则使用当前坐标轴。
- `diagonal`: 对角线上的单变量分布图，可以选择 'hist' 或 'kde'。
- `marker`: 散点的标记样式，默认为 '.'，还可以使用其他的标记样式，例如 "."、"+" 等。
- `density_kwds`: 字典类型，传递给核密度估计图(kde)的参数设置。
- `hist_kwds`: 字典类型，传递给直方图(hist)的参数设置。
- `range_padding`: 控制图像中每个小图与边界的距离。
- `**kwargs`: 其他关键字参数，比如颜色等。具体有
    - `c`: 用于指定散点的颜色，可以是单一整数或字符串，也可以通过传递列表来指定每个数据集的颜色。
    - `s`: 用于指定散点的大小，默认为 20。
    - `cmap`: 用于指定颜色映射的名称，默认为 "viridis"。
    - `norm`: 用于归一化颜色数据的方法。
    - `vmin`, `vmax`: 用于指定颜色显示的最小值和最大值。
    - `linewidths`: 用于指定散点边界的线宽，默认为 None。
    - `edgecolors`: 用于指定散点边界的颜色，默认为 None。
    - `label`: 可以是字符串或数组，包含每个小图的标签名称。


````python
from pandas.plotting import scatter_matrix
import pandas as pd
import numpy as np
````

````python
# 创建一个 DataFrame 对象，包含两个特征 'x' 和 'y'
data = pd.DataFrame(np.random.randn(100, 2), columns=['x', 'y'])

# 使用 scatter_matrix 函数创建散点图矩阵
scatter_matrix(data, alpha=0.2, figsize=(6, 6), diagonal='hist')

plt.show()
````
