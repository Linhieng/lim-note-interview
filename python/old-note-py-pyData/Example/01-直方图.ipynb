````python
import matplotlib.pyplot as plt
import numpy as np
````

## 1. 简单使用

直方图(Histogram), 一般用于描述连续型数据的分布情况。 它将数据划分成若干个等距的区间，并用每个区间的高度来表示该区间内的数据数量或频率。

而 bar 一般用于描述离散型数据的大小或数量。 它将每个数据按照类别进行分组。

````python
# 生成一些随机数据
data = np.random.randn(1000)

# 绘制直方图
plt.hist(data)

# 显示图形
plt.show()
````

````python
# 生成一些随机数据
data = np.random.randn(1000)

# 绘制直方图
plt.hist(data, bins=50, histtype='bar', linewidth=1, edgecolor='black')

plt.ylim(0,70)
plt.xlim(-5,5)

# 显示图形
plt.show()
````

## 2. 相关参数

语法:
```py
matplotlib.pyplot.hist(x, bins=None, range=None, density=None, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)
```
- `x`: (array_like) 要绘制直方图的数据
- `bins`: (int, str, sequence, optional) 直方图条形数目。可以是一个整数，表示bins数量，也可以是一个序列，定义每个bin的边界值。默认值为10。
- `range`: (tuple, optional) 数据显示范围，将会把所有数据限制在这个范围内。默认使用data的范围。
- `density`: (bool, optional) 是否显示概率密度而不是计数，默认值为False。
- `weights`: (array_like, optional) 与x长度相同的数组。这些数组用于加权x中的值。默认情况下，每个值都具有相同的重量。
- `cumulative`: (bool, optional) 是否要对每个元素的累加进行直方图处理。如果设置为True，则返回累积直方图，默认为False。
- `bottom`: (scalar, optional) 直方图的y坐标轴最低点位置。默认值为None，即从0开始。
- `histtype`: ('bar', 'barstacked', 'step', 'stepfilled') 直方图类型。根据需要选择，例如：'bar'，'barstacked'，'step'，'stepfilled'。默认是'bar'。
- `align`: ('left', 'mid', 'right'), optional) 条形边缘的位置。默认是'left'。
- `orientation`: ('horizontal', 'vertical') 直方图定向，默认为'vertical'。
- `rwidth`: (scalar, optional) 直方图条形宽度，相对于bin宽度的比例。如果不提供，则自动计算。无法与多项式或步幅虚线histtypes一起使用。 如果cumulative=True，则忽略此选项并保持基本宽度为1。
- `log`: (bool, optional) 是否对y轴进行对数缩放。默认为False。
- `color`: (color, optional) 直方图颜色。
- `label`: (str, optional) 图例标签。
- `stacked`: (bool, optional) 是否堆叠直方图。默认为False。
- `normed`: (deprecated, use density instead!)，与density参数等效。
- `data`: (array or DataFrame or list-like) 作为输入数据的对象。
