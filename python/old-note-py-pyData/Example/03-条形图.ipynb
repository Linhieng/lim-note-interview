```python
import matplotlib.pyplot as plt
```

## 1. 简单使用

条形图, 一般指的是横着放的图表, 竖着放的叫做柱形图。

```python
# 数据
labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 23, 45, 32, 18]

# 绘制条形图
plt.barh(labels, values)

# 添加标题和标签
plt.title('Bar Chart Horizontal')
plt.xlabel('Category')
plt.ylabel('Value')

# 显示图表
plt.show()
```
