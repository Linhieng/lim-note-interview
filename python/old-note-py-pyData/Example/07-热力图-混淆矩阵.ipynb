`annot` 控制是否在矩阵中显示数值

`fmt` 控制数值的格式, 比如 `'d'` 就是显示为整数。

```python
# Generate a sample confusion matrix
cm = [[100, 20, 30], [40, 80, 10], [5, 25, 90]]

# Define color map
cmap = 'Blues'

# Create heatmap with annotated values
sns.heatmap(cm, annot=True, fmt='d', cmap=cmap)

# Show plot
plt.show()
```

```python
cm = [[100, 20, 30], [40, 80, 10], [5, 25, 90]]

# 绘制热力图（Matplotlib）, plt 没有 annot 属性来显示数值, , 不过它也有 cmap 属性
plt.matshow(cm, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
# 创建 3x3 的随机矩阵列表
matrices = [np.random.rand(3, 3) for _ in range(9)]

j = 1
fig = plt.figure(figsize=(12, 12))
for i in range(len(matrices)):
    plt.subplot(3, 3, j)
    j += 1
    sns.heatmap(matrices[i], annot=True, cmap='Blues')

# 调整子图之间的间距
fig.subplots_adjust(hspace=0.4, wspace=0.4)
plt.show()
```
