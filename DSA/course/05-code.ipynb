# [搜索二叉树](https://www.nowcoder.com/practice/a69242b39baf45dea217815c7dedb52b)

默认传入的 `root` 不是 `None`

```python
def is_valid_BST1(root):
    # 存储中序遍历的结果
    inorder_list = []
    inorder(root, inorder_list)
    for i in range(1, len(inorder_list)):
        if inorder_list[i] <= inorder_list[i-1]:
            return False
    return True


def inorder(root, inorder_list):
    if root is None:
        return
    inorder(root.left, inorder_list)
    inorder_list.append(root.val)
    inorder(root.right, inorder_list)
```

```python
def is_valid_BST2(root):
    # 传递地址
    pre_value = [None]
    return process(root, pre_value)


def process(root, pre_value):
    if root is None:
        return True
    # 先判断左侧是不是 BST
    is_left_BST = process(root.left, pre_value)
    if False == is_left_BST:
        return False

    # 判断是否升序
    if pre_value[0] is not None and pre_value[0] >= root.val:
        return False
    else:
        pre_value[0] = root.val

    # 接着判断右侧是否是 BST
    return process(root.right, pre_value)
```

```python
def is_valid_BST3(root):
    stack = []
    p = root
    pre_value = None
    while len(stack) != 0 or p is not None:
        if p is not None:
            stack.append(p)
            p = p.left
        else:
            p = stack.pop()
            # 判断是否升序
            if pre_value is not None and pre_value >= p.val:
                return False
            else:
                pre_value = p.val
            p = p.right
    return True
```

```python
def is_valid_BST4(root):
    return process(root)[0]


def process(x):
    if x is None:
        return None

    x_max = x_min = x.val

    l = process(x.left)
    if l is not None:
        l_is_BST, l_max, l_min = l
        if False == l_is_BST or l_max >= x.val:
            return (False, None, None)
        x_max = max(x_max, l_max)
        x_min = min(x_min, l_min)

    r = process(x.right)
    if r is not None:
        r_is_BST, r_max, r_min = r
        if False == r_is_BST or x.val >= r_min:
            return (False, None, None)
        x_max = max(x_max, r_max)
        x_min = min(x_min, r_min)

    return (True, x_max, x_min)
```

# [完全二叉树](https://www.nowcoder.com/practice/8daa4dff9e36409abba2adbe413d6fae)

```python
from queue import Queue


def is_complete_tree(root):
    queue = Queue()
    queue.put(root)
    check_leaf_flag = False
    while not queue.empty():
        p = queue.get()
        l = p.left
        r = p.right

        if (
            (r is not None and l is None)  # 条件1
            or
            (check_leaf_flag and (l is not None or r is not None))  # 条件2
        ):
            return False

        queue.put(l) if l is not None else None
        queue.put(r) if r is not None else None

        if r is None:
            check_leaf_flag = True
    return True
```

# TEST - 满二叉树

## 工具

```python
from queue import Queue
from random import randint as r
from time import time as t


class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None


def generate_full_binary_tree(deep=3, max_value=10):
    p = root = TreeNode(1)
    stack = []
    stack.append(p)
    while len(stack) != 0:
        p = stack[-1]
        if len(stack) >= deep:
            # 达到深度, 说明是叶节点, 可以出栈
            stack.pop()
        elif p.left is None:
            # 没有左子节点, 则创建左子节点
            p.left = TreeNode(1)
            stack.append(p.left)
        elif p.right is None:
            # 没有右子节点, 则创建右子节点
            p.right = TreeNode(1)
            stack.append(p.right)
        else:
            # 不是叶节点, 同时有两个子节点, 可以出栈
            stack.pop()
    return root


def generate_random_binary_tree(deep=3, max_value=10):
    p = root = TreeNode(2)
    stack = []
    stack.append(p)
    while len(stack) != 0:
        p = stack[-1]
        if r(0, 9) == 1 and len(stack) >= deep:
            # 达到深度, 说明是叶节点, 可以出栈
            stack.pop()
        elif r(0, 1) == 1 and p.left is None:
            # 没有左子节点, 则创建左子节点
            p.left = TreeNode(2)
            stack.append(p.left)
        elif r(0, 1) == 1 and p.right is None:
            # 没有右子节点, 则创建右子节点
            p.right = TreeNode(2)
            stack.append(p.right)
        else:
            # 不是叶节点, 同时有两个子节点, 可以出栈。 或者"中奖"了, 也可以出栈
            stack.pop()
    return root


def generate_binary_tree(max_deep=10, max_value=20):
    # 不生成空二叉树
    deep = r(1, max_deep)
    if r(0, 5) < 1:
        # 生成满二叉树
        return generate_full_binary_tree(deep, max_value)
    else:
        # 随机生成二叉树
        return generate_random_binary_tree(deep, max_value)
```

## 实现

```python
def is_full_binary_tree1(root):
    nodes, deep = get_nodes_deep(root)
    return nodes == ((1 << deep) - 1)

def get_nodes_deep(root):
    if root is None:
        return 0, 0
    l_nodes, l_deep = get_nodes_deep(root.left)
    r_nodes, r_deep = get_nodes_deep(root.right)
    nodes = 1 + l_nodes + r_nodes 
    deep = 1 + max(r_deep, l_deep) 
    return nodes, deep
```

```python
def is_full_binary_tree2(root):
    queue = Queue()
    queue.put(root)

    deep = 0
    nodes = 0
    cur_end = root
    next_end = None

    while not queue.empty():
        p = queue.get()
        nodes += 1
        if (p.left is None and p.right is not None) or (p.left is not None and p.right is None):
            return False
        if p.left is not None:
            next_end = p.left
            queue.put(p.left)
        if p.right is not None:
            next_end = p.right
            queue.put(p.right)
        if p is cur_end:
            cur_end = next_end
            next_end = None
            deep += 1
    return nodes == (1<<deep)-1
```

## 测试

```python
test_time, max_size, max_value, succeed = 10_000, 10, 1000, True

st = t()
for _ in range(test_time):
    root = generate_binary_tree(max_size, max_value)
    test_flag1 = is_full_binary_tree1(root)
    test_flag2 = is_full_binary_tree2(root)
    if test_flag1 != test_flag2:
        succeed = False
        break
print(f'{t()-st:5.2f}s, 判断满二叉树: ', '✔️' if succeed else '❌')
```

# [平衡二叉树(AVL 树)](https://www.nowcoder.com/practice/8b3b95850edb4115918ecebdf1b4d222)

```python
"""
- 是否是平衡二叉树
- 高度
"""
def is_balanced_binary_tree(root):
    return is_balanced(root)[0]

def is_balanced(head):
    if head is None:
        return (True, 0)
    l_is_AVL, l_height = is_balanced(head.left)
    if False == l_is_AVL:
        return (False, None)
    
    r_is_AVL, r_height = is_balanced(head.right)
    if False == r_is_AVL:
        return (False, None)
    
    is_AVL = True
    height = 1 + max(l_height, r_height)
    if abs(l_height - r_height) > 1:
        is_AVL = False
    
    return (is_AVL, height)
```

# [最近公共祖先](https://www.nowcoder.com/practice/e0cc33a83afe4530bcec46eba3325116)

```python
# o1_val o2_val 是 root 树上两个节点的值。 题设保证二叉树中每个结点的值都不相同
def lowest_common_ancestor1(root, o1_val, o2_val):
    if root.val == o1_val or root.val == o2_val:
        return root.val

    parent_val_map = {}
    parent_val_map[root.val] = root.val
    init_parent_val_map(root, parent_val_map)

    ancestor_set = set()
    while o1_val != root.val:
        ancestor_set.add(o1_val)
        o1_val = parent_val_map[o1_val]

    while o2_val != root.val:
        if o2_val in ancestor_set:
            return o2_val
        o2_val = parent_val_map[o2_val]

    return root.val


def init_parent_val_map(head, parent_val_map):
    if head is None:
        return
    if head.left is not None:
        parent_val_map[head.left.val] = head.val
        init_parent_val_map(head.left, parent_val_map)
    if head.right is not None:
        parent_val_map[head.right.val] = head.val
        init_parent_val_map(head.right, parent_val_map)
```

```python
# o1_val o2_val 是 root 树上两个节点的值。 题设保证二叉树中每个结点的值都不相同
def lowest_common_ancestor2(root, o1_val, o2_val):
    if root.val == o1_val or root.val == o2_val:
        return root.val

    ancestor_set = set()
    get_ancestor_set(root, o1_val, ancestor_set)
    
    parent_val_map = {}
    parent_val_map[root.val] = root.val
    init_parent_val_map(root, parent_val_map)

    while o2_val != root.val:
        if o2_val in ancestor_set:
            return o2_val
        o2_val = parent_val_map[o2_val]

    return root.val


def get_ancestor_set(root, node_val, ancestor_set):
    if root is None:
        return False
    elif root.val == node_val:
        return True

    if True == get_ancestor_set(root.left, node_val, ancestor_set):
        ancestor_set.add(root.val)
        return True
    if True == get_ancestor_set(root.right, node_val, ancestor_set):
        ancestor_set.add(root.val)
        return True
    
    return False
```

```python
def lowest_common_ancestor3(head, o1_val, o2_val):
    if head.val == o1_val or head.val == o2_val:
        return head.val
    left, right = -1, -1
    if head.left is not None:
        left = lowest_common_ancestor3(head.left, o1_val, o2_val)
    if head.right is not None:
        right = lowest_common_ancestor3(head.right, o1_val, o2_val)
    if left != -1 and right != -1:
        return head.val
    return left if left != -1 else right
```

```python
# 保证传入的二叉树是二叉搜索树, 即左小右大
def lowest_common_ancestor_BST(head, o1_val, o2_val):
    while head is not None:
        if head.val > o1_val and head.val > o2_val:  # o1 o2 在左边
            head = head.left  
        elif head.val < o1_val and head.val < o2_val:  # o1 o2 在右边
            head = head.right
        else: # o1 o2 在两侧
            break
    return head.val
```

# [后继节点](https://www.nowcoder.com/practice/9023a0c988684a53960365b889ceaf5e)

```python
"""
传入的 node 右三种情况:
初次思考时, 会认为 node 有三种情况: 是左子节点, 父节点, 右子节点。
但实际上, 只有根节点是纯粹的父节点外, 其他位置的节点, 都是父节点+左或右子节点
所以考虑父节点时, 只需要思考根节点的情况;
考虑左子节点时, 还要考虑他是父节点;
同理考虑右子节点时, 也要考虑他是父节点。
因为我们单独思考了根节点, 所以考虑左右子结点时不需要考虑左右子结点没有父节点的情况。

所以应该这样分析:
- node 是根节点, 下一节点就是右子树的最左侧叶节点
- node 是左子节点, 下一节点就是父节点。
- node 是右子节点
    - 如果父节点是左子节点, 则下一节点是父节点的父节点
    - 如果父节点是右子节点, 还得继续往上找 (注意此时不要立马想着下一节点是空)
        - 如果父节点往上能找到作为左子节点存在的节点, 则下一节点是该节点的父节点
        - 如果往上一直找, 发现节点一直都是右子节点, 说明 node 是最右侧的节点, 下一节点为空

整理上面的情况:
- 如果 node 右子节点不为空, 则下一节点就是右子节点的最左侧叶节点
- 如果为空, 则往上找, 找到一个节点(可以是自己)作为左子节点的存在, 下一节点就是该左子节点的父节点。 找不到则说明下一节点为空
"""
def get_next_node_inorder(node):
    if node.right is not None:
        node = node.right
        while node.left is not None:
            node = node.left
        return node
    while node.next is not None and node is not node.next.left:  # 判断当前节点是否是左子节点
        node = node.next
    return node.next
```

# [二叉树的序列化和反序列化](https://www.nowcoder.com/practice/cf7e25aa97c04cc1a68c8f040e71fb84)

```python
from queue import Queue


class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

## 先序

```python
def preorder_serialize_recursion(head):
    if head is None:
        return '#_'
    s = f'{head.val}_'
    s += preorder_serialize_recursion(head.left)
    s += preorder_serialize_recursion(head.right)
    return s


def preorder_deserialize_recursion(s):
    if s is None:
        return None
    queue = Queue()
    nodes = s.split('_')
    nodes.pop()
    for node in nodes:
        queue.put(node)
    return preorder_deserialize_recursion_helper(queue)


def preorder_deserialize_recursion_helper(queue):
    val = queue.get()
    if val == '#':
        return None
    head = TreeNode(int(val))
    head.left = preorder_deserialize_recursion_helper(queue)
    head.right = preorder_deserialize_recursion_helper(queue)
    return head
```

```python
def preorder_serialize(root):
    if root is None:
        return None
    def _node(node):
        return f'{int(node.val)}_' if node is not None else '#_'
    
    stack = [root]
    s = ''
    while 0 != len(stack):
        head = stack.pop()

        if head is None:
            s += '#_'
        else:
            s += _node(head)
            stack.append(head.right)
            stack.append(head.left)
    return s

def preorder_deserialize_recursion(s):
    if s is None or s == '#_':
        return None
    nodes = s.split('_')
    t = [0]
    return preorder_deserialize_recursion_helper(nodes, t)

def preorder_deserialize_recursion_helper(nodes, t):
    if nodes[t[0]] == '#':
        return None
    head = TreeNode(int(nodes[t[0]]))
    t[0] += 1
    head.left = preorder_deserialize_recursion_helper(nodes, t)
    t[0] += 1
    head.right = preorder_deserialize_recursion_helper(nodes, t)
    return head
    
```

## 后序

```python
def postorder_serialize_recursion(head):
    if head is None:
        return '#_'
    s = postorder_serialize_recursion(head.left)
    s += postorder_serialize_recursion(head.right)
    s += f'{head.val}_'
    return s

def postorder_deserialize_recursion(s):
    if s is None:
        return None
    stack = []
    nodes = s.split('_')
    nodes.pop()
    for node in nodes:
        stack.append(node)
    return postorder_deserialize_recursion_helper(stack)

def postorder_deserialize_recursion_helper(stack):
    val = stack.pop()
    if val == '#':
        return None
    head = TreeNode(int(val))
    head.right = postorder_deserialize_recursion_helper(stack)
    head.left = postorder_deserialize_recursion_helper(stack)
    return head
```

## 层序

```python
def traverse_serialize(root):
    if root is None:
        return '#_'
    queue = Queue()
    queue.put(root)
    s = f'{root.val}_'
    while not queue.empty():
        p = queue.get()

        if p.left is not None:
            queue.put(p.left)
            s += f'{p.left.val}_'
        else:
            s += f'#_'
        if p.right is not None:
            queue.put(p.right)
            s += f'{p.right.val}_'
        else:
            s += f'#_'
    return s


def traverse_deserialize(s):
    if s is None or s == '#_':
        return None
    def _node(val):
        return TreeNode(int(val)) if val != '#' else None
    
    # nodes 中从第一个元素(包含该元素)开始, 每两个元素是某一节点的两个子节点。
    # 根据这个规律, 可以按层序依次生成节点, 然后得到它的两个子节点
    nodes = s.split('_')
    i = 0
    root = head = _node(nodes[i])
    i += 1
    # queue 中按层序依次添加新节点。
    queue = Queue()
    queue.put(head)
    while not queue.empty():
        head = queue.get()
        head.left, head.right = _node(nodes[i]), _node(nodes[i+1])
        i += 2
        if head.left is not None:
            queue.put(head.left)
        if head.right is not None:
            queue.put(head.right)

    return root
```
